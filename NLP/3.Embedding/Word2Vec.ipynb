```python
import os
import re
import numpy as np
import pandas as pd
import nltk
import warnings
import gensim

from nltk import sent_tokenize
from gensim.utils import simple_preprocess

warnings.filterwarnings('ignore')
nltk.download('all')


from pathlib import Path
script_dir = Path().resolve()
csv_path = os.path.join(script_dir, "..", "Resource", "Reviews.csv")

df = pd.read_csv(csv_path, nrows=100)
```

```python
raw_sentences = df["Text"].dropna().astype(str).tolist()
raw_sentences
```

```python
# Tokenize the sentences
tokenize_sentences = [sent_tokenize(sentence) for sentence in raw_sentences]
tokenize_sentences
```

```python
all_words = []
for sentence in raw_sentences:
    all_words.append(simple_preprocess(sentence))
```

```python
 all_words
 len(all_words)
```

```python
# ----------- Word2Vec Model Training (Skip-gra) # sg=1-------------
custom_model_SP = gensim.models.Word2Vec(
    sentences=raw_sentences,
    vector_size=150,
    window=10,
    min_count=5,  
    sg=1
)
custom_model_SP.build_vocab(all_words, update=True) 

len(custom_model_SP.wv.index_to_key)
```

```python
 all_words
```

```python
custom_model_SP.train(raw_sentences, total_examples=custom_model_SP.corpus_count, epochs=20)
```
